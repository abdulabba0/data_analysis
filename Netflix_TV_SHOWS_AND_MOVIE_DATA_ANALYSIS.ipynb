## **Handle Missing Data**

# replace NA in director column with no director

```python
df['director'] = df['director'].fillna('No Director')
```

```python
df.info()
```

### remove missing columns

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
df['num_cast'] = df.cast.str.split(',').str.len().astype('Int64')
```

```python
df[['title', 'num_cast']]
```

### check the maximum num of cast per movie

### Plotting value count for cast

```python
len(df[df['num_cast'] <= 21])
```

```python
cast_value_count = df[df['num_cast'] ]['num_cast'].value_counts().sort_index().plot(kind='barh')
plt.xlabel("Cast per Movie")
plt.ylabel("Count")
plt.title("Value Count of Cast per Movie")
plt.tight_layout()
plt.show()
```
