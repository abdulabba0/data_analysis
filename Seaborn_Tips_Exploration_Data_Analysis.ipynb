##Data Visualization

```python
# Plotting a bar graph for the sex
plt.figure(figsize=(5,4))
plt.title('Sex Distribution')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.bar(data['sex'].value_counts().index, data['sex'].value_counts())
plt.show()
```

```python
# Plotting distribution of smoker
plt.figure(figsize=(5,4))
plt.title('Smoker Distribution')
plt.xlabel('Smoker')
plt.ylabel('Count')
plt.pie(data['smoker'].value_counts().values, labels=data.smoker.value_counts().index, autopct='%1.1f%%')
plt.show()
```

```python
# Plot Distribution for day
plt.figure(figsize=(5,4))
plt.title('Day Distribution')
plt.xlabel('Day')
plt.ylabel('Count')
plt.bar(data['day'].value_counts().index, data['day'].value_counts())
plt.show()
```

```python
# Pie plot for day distribution
plt.figure(figsize=(5,4))
plt.title('Day Distribution')
plt.xlabel('Day')
plt.ylabel('Count')
plt.pie(data['day'].value_counts().values, labels=data.day.value_counts().index, autopct='%1.1f%%')
plt.show()
```

```python
# plot distribution for time
plt.figure(figsize=(5,4))
plt.title('Time Distribution')
plt.xlabel('Time')
plt.ylabel('Count')
plt.pie(data['time'].value_counts().values, labels=data.time.value_counts().index, autopct='%1.1f%%')
plt.show()
plt.show()
```

```python
# Plot for size distribution
plt.figure(figsize=(5,4))
plt.title('Size Distribution')
plt.xlabel('Size')
plt.ylabel('Count')
plt.pie(data['size'].value_counts().values, labels=data.size.value_counts().index, autopct='%1.1f%%')
plt.show()
# plt.bar(data['size'].value_counts().index, data['size'].value_counts())

```

```python
# Plot for tip
plt.figure(figsize=(5,4))
plt.title('Tip Distribution')
plt.xlabel('Tip')
plt.ylabel('Count')
plt.hist(data['tip'])
plt.show()
```

```python
# data.value_counts("sex")
male_percentage = (data['sex'].value_counts()['Male'] / len(data)) * 100
female_percentage = (data['sex'].value_counts()['Female'] / len(data)) * 100
print(f"Male Percentage: {male_percentage:.2f}%")
print(f"Female Percentage: {female_percentage:.2f}%")
```

```python
data['tip_percentage'] = (data['tip'] / data['total_bill'])*100
data['tip_percentage']
```
