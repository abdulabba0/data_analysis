##**What is Matplotlib**

Matplotlib is a widely used plotting library for the Python programming language. It provides a variety of tools for creating static, interactive, and animated visualizations in Python. Here are some key points about Matplotlib:

**Key Features of Matplotlib**
- Wide Range of Plots: Supports various types of plots such as line plots, bar charts, histograms, scatter plots, pie charts, error bars, box plots, and more.
- Customization: Highly customizable, allowing users to adjust colors, labels, legends, and other plot elements.
- Integration: Can be used with other Python libraries like NumPy, Pandas, and SciPy for data manipulation and analysis.
- Interactivity: Can create interactive plots with tools like Jupyter notebooks, enabling zooming, panning, and updating plots in real-time.
- Publication Quality: Capable of producing high-quality, publication-ready graphics in various formats (PNG, PDF, SVG, etc.).
- Subplotting: Allows for the creation of complex layouts with multiple plots in a single figure.

###**Let's see how all the plot looks like**

```python
#== ALL PLOTS IN MATPLOTLIB ===


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate some example 3D data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Create a figure with subplots
fig = plt.figure(figsize=(18, 18))

# Plotting different types of plots

# Line plot
ax1 = fig.add_subplot(4, 4, 1)
ax1.plot(x[0, :], z[0, :], color='blue')
ax1.set_title('Line Plot')

# Bar plot
ax2 = fig.add_subplot(4, 4, 2)
ax2.bar(['A', 'B', 'C', 'D'], [3, 7, 2, 5], color='green')
ax2.set_title('Bar Plot')

# Histogram
ax3 = fig.add_subplot(4, 4, 3)
ax3.hist(np.random.normal(size=100), bins=20, color='orange')
ax3.set_title('Histogram')

# Scatter plot
ax4 = fig.add_subplot(4, 4, 4)
ax4.scatter(np.random.normal(size=100),
np.random.normal(size=100), color='red')
ax4.set_title('Scatter Plot')

# Box plot
ax5 = fig.add_subplot(4, 4, 5)
ax5.boxplot(np.random.normal(size=100), vert=False,
patch_artist=True, meanline=True, showmeans=True)
ax5.set_title('Box Plot')

# Contour plot
ax6 = fig.add_subplot(4, 4, 6)
ax6.contourf(x, y, z, cmap='viridis')
ax6.set_title('Contour Plot')

# Image Plot
ax7 = fig.add_subplot(4, 4, 7)
ax7.imshow(z, cmap='plasma', extent=(-5, 5, -5, 5))
ax7.set_title('Image Plot')

# 3D Surface Plot
ax8 = fig.add_subplot(4, 4, 8, projection='3d')
ax8.plot_surface(x, y, z, cmap='coolwarm')
ax8.set_title('3D Surface Plot')

# Horizontal Barplot
ax9 = fig.add_subplot(4, 4, 9)
ax9.barh(['Category1', 'Category2', 'Category3'],
[5, 3, 7], color='purple')
ax9.set_title('Horizontal Bar Plot')

# Violin Plot
ax10 = fig.add_subplot(4, 4, 10)
ax10.violinplot(np.random.normal(size=100), vert=False)
ax10.set_title('Violin Plot')

# Stem Plot
ax11 = fig.add_subplot(4, 4, 11)
ax11.stem(x[0, :], z[0, :], linefmt='b-', markerfmt='bo', basefmt='r-')
ax11.set_title('Stem Plot')

# 3D Scatter Plot
ax12 = fig.add_subplot(4, 4, 12, projection='3d')
ax12.scatter(np.random.normal(size=100), np.random.normal(size=100),
np.random.normal(size=100), color='green')
ax12.set_title('3D Scatter Plot')

# Pie Plot
ax13 = fig.add_subplot(4, 4, 13)
ax13.pie([15, 30, 45, 10], labels=['A', 'B', 'C', 'D'],
autopct='%1.1f%%',
colors=['gold', 'yellowgreen', 'lightcoral', 'lightskyblue'])
ax13.set_title('Pie Plot')

# Step Plot
ax14 = fig.add_subplot(4, 4, 14)
ax14.step(x[0, :], np.cos(x[0, :]),
color='magenta')
ax14.set_title('Step Plot')

# Fill Polygon Plot
ax15 = fig.add_subplot(4, 4, 15)
ax15.fill(x[0, :], np.tan(x[0, :]),
color='cyan', alpha=0.3)
ax15.set_title('Fill Polygon Plot')

# Bar3D Plot
ax16 = fig.add_subplot(4, 4, 16, projection='3d')
x_data = np.arange(5)
y_data = np.arange(5)
x_data, y_data = np.meshgrid(x_data, y_data)
z_data = np.random.rand(5, 5) * 10
ax16.bar3d(x_data.ravel(), y_data.ravel(),
np.zeros_like(z_data.ravel()), 1, 1, z_data.ravel(), shade=True)
ax16.set_title('Bar3D Plot')

# Adjust layout
plt.tight_layout()

# Save the plot to a file
plt.savefig("matplotlib.png")

# Show the plot
plt.show()
```

```python
# @title
# Generate some example data
x = np.linspace(0, 10, 50)
y = np.sin(x)

# List of matplotlib markers
markers = [
    '.', ',', 'o', 'v', '^', '<', '>', '1', '2', '3', '4', 's', 'p', '*', 'h', 'H',
    '+', 'x', 'D', 'd', '|', '_'
]

# Create a figure with subplots
fig, axs = plt.subplots(5, 5, figsize=(20, 20))

# Plot using each marker
for i, marker in enumerate(markers):
    row = i // 5
    col = i % 5
    ax = axs[row, col]
    ax.plot(x, y, marker=marker, linestyle='None', markersize=10)
    ax.set_title(f'Marker: {marker}')
    ax.grid(True)

# Adjust layout
plt.tight_layout()
plt.savefig('plot.png')

# Show the plot
plt.show()
```

##**Customizing Labels and Titles in Matplotlib**

> Matplotlib allows extensive customization of plot titles, axis labels, and tick labels. You can adjust their size, positioning, color, and font to enhance the readability and aesthetics of your plots. Below is an explanation of how to customize these elements:

1. **Title Customization**
- Font Size: Use fontsize to change the size of the title text.
- Color: Use color to set the color of the title text.
- Font Weight: Use fontweight to make the title text bold, light, etc.
- Positioning: Use loc to position the title ('left', 'center', 'right').
- Padding: Use pad to add space between the title and the plot.

2. **Axis Label Customization**
- Font Size: Use fontsize to change the size of the axis label text.
Color: Use color to set the color of the axis label text.
- Font Style: Use fontstyle to italicize the label text.
Padding: Use labelpad to add space between the label and the axis.

3. **Tick Label Customization**
- Font Size: Use labelsize to change the size of the tick label text.
- Color: Use labelcolor to set the color of the tick label text.
Rotation: Use rotation to rotate the tick labels.

4. **Adding a Grid**
Use ax.grid(True) to add a grid to the plot for better readability.

```python
# Generate some example data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a figure and an axis
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
ax.plot(x, y)

# Customize the title
ax.set_title('Sine Wave', fontsize=20, color='purple', fontweight='bold', loc='center', pad=20)

# Customize the x and y labels
ax.set_xlabel('X-axis', fontsize=15, color='blue', fontstyle='italic', labelpad=10)
ax.set_ylabel('Y-axis', fontsize=15, color='green', fontstyle='italic', labelpad=10)

# Customize tick labels
ax.tick_params(axis='x', labelsize=12, labelcolor='red', rotation=45)
ax.tick_params(axis='y', labelsize=12, labelcolor='brown', rotation=45)

# Add grid
ax.grid(True)

# Display the plot
plt.tight_layout()
plt.show()
```

##**Matplotlib Subplots**

> Matplotlib provides a convenient way to create multiple plots within a single figure using the subplot function. This is useful for comparing different datasets or visualizing different aspects of the same data side by side.

> What subplot Does
The subplot function in Matplotlib creates a grid of subplots within a single figure. Each subplot is a separate axes object where you can plot different data. The subplot function can be used in different ways to create subplots in various configurations.

####**How to Use subplot**
Basic Usage:

- plt.subplot(nrows, ncols, index): Creates a grid of subplots with nrows rows and ncols columns. The index specifies the position of the subplot in the grid (starting from 1 in the top-left corner and increasing row-wise).

**Using plt.subplots():**

- plt.subplots(nrows, ncols): Returns a tuple containing the figure and an array of Axes objects. This is a more flexible way to create subplots, allowing easy iteration over the axes.
Adding Titles to Each Subplot
You can add titles to each subplot using the set_title method of the Axes object.

```python
# Generate some example data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(x)

# Create a figure and a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# Plot on each subplot and set titles
axs[0, 0].plot(x, y1, color='b')
axs[0, 0].set_title('Sine Wave')

axs[0, 1].plot(x, y2, color='g')
axs[0, 1].set_title('Cosine Wave')

axs[1, 0].plot(x, y3, color='r')
axs[1, 0].set_title('Tangent Wave')

axs[1, 1].plot(x, y4, color='m')
axs[1, 1].set_title('Exponential Function')

# Adjust layout to prevent overlap
plt.tight_layout()
# Create a title for the entire plot
plt.suptitle('Subplots Example', fontsize=18, fontweight='bold', color="brown")

# Adding padding to top between top of the sublot and the figure, same as
# bottom, left and right
plt.subplots_adjust(top=0.9, bottom=0.1, left=0.1,
                    right=0.9, hspace=0.4, wspace=0.4)

# Show the plot
plt.show()
```

##**Types of Graphs in Matplotlib**

Matplotlib provides various plotting options to visualize data. Each type of graph is suited to different kinds of data and datasets. Below are detailed explanations and code snippets for common types of graphs, including their uses, suitable datasets, and types of data they are best for.

###**Line Plot**
- Use: To display trends over time or continuous data.
- Suitable Dataset: Time series data (e.g., stock prices, temperature over time).
- Type of Data: Numerical.
Example Data: Daily closing prices of a stock.

```python
# Example data: Daily closing prices of a stock over 100 days
days = np.arange(1, 101)
closing_prices = np.random.normal(100, 10, size=100)

plt.figure(figsize=(10, 5))
plt.plot(days, closing_prices, marker='o')
plt.title('Daily Closing Prices of Stock')
plt.xlabel('Day')
plt.ylabel('Closing Price ($)')
plt.grid(True)
plt.show()
```

###**Bar Plot**
- Use: To compare different categories.
- Suitable Dataset: Categorical data (e.g., sales per region, number of students in different classes).
- Type of Data: Categorical and numerical.
- Example Data: Sales figures for different regions.

```python
# Example data: Sales figures for different regions
regions = ['North', 'South', 'East', 'West']
sales = [250, 400, 350, 300]

plt.figure(figsize=(10, 5))
plt.bar(regions, sales, color='skyblue')
plt.title('Sales Figures by Region')
plt.xlabel('Region')
plt.ylabel('Sales ($)')
plt.show()
```

###**Histogram**
- Use: To show the distribution of a dataset.
- Suitable Dataset: Data with continuous or discrete variables (e.g., exam scores, heights of people).
- Type of Data: Numerical.
- Example Data: Heights of 1000 people.

```python
# Example data: Heights of 1000 people
heights = np.random.normal(170, 10, size=1000)

plt.figure(figsize=(10, 5))
plt.hist(heights, bins=30, color='green', edgecolor='black')
plt.title('Distribution of Heights')
plt.xlabel('Height (cm)')
plt.ylabel('Frequency')
plt.show()
```

**Scatter Plot**
- Use: To display the relationship between two variables.
- Suitable Dataset: Data with two continuous variables (e.g., height vs. weight, age vs. income).
- Type of Data: Numerical.

###**Customizing Scatter Plots in Matplotlib**

> Scatter plots are useful for visualizing the relationship between two variables. Matplotlib provides several customization options to enhance the visualization of scatter plots, including changing the size of the dots, their color, adding colormaps, and introducing transparency.

1. Changing the Size of the Dots
The size of the dots can be changed using the s parameter in the scatter function.
This parameter accepts either a single scalar value to set a uniform size for all dots or an array of values to set individual sizes.
2. Changing the Color of the Dots
The color of the dots can be changed using the color or c parameter.
This parameter can accept a single color value, an array of color values, or a colormap.
3. Adding Colormap
A colormap can be used to map data values to colors, enhancing the visual representation.
The c parameter can accept an array of values, and the cmap parameter specifies the colormap.
4. Adding Transparency
The transparency of the dots can be adjusted using the alpha parameter.
This parameter accepts a value between 0 (fully transparent) and 1 (fully opaque).

```python
# Generate example data
np.random.seed(42)
x = np.random.rand(100)
y = np.random.rand(100)
sizes = 1000 * np.random.rand(100)  # Random sizes for the dots
colors = np.random.rand(100)  # Random colors for the dots

fig, axs = plt.subplots(2, 2, figsize=(15, 10))

# Default scatter plot
axs[0, 0].scatter(x, y)
axs[0, 0].set_title('Default Scatter Plot')

# Scatter plot with custom sizes
axs[0, 1].scatter(x, y, s=sizes, color='blue', alpha=0.6)
axs[0, 1].set_title('Scatter Plot with Custom Sizes')

# Scatter plot with colormap
scatter = axs[1, 0].scatter(x, y, c=colors, cmap='viridis', alpha=0.6)
fig.colorbar(scatter, ax=axs[1, 0])
axs[1, 0].set_title('Scatter Plot with Colormap')

# Scatter plot with transparency
axs[1, 1].scatter(x, y, s=100, color='red', alpha=0.3)
axs[1, 1].set_title('Scatter Plot with Transparency')

plt.tight_layout()
plt.show()
```

###**Understanding Relationships Between Variables Using Scatter Plots**

Scatter plots are a powerful visualization tool used to understand the relationship between two continuous variables. Below are some common types of relationships that can be identified using scatter plots, along with code examples to visualize them:

####**Positive Linear Relationship**

- In a positive linear relationship, as one variable increases, the other variable also increases. This relationship is depicted as a set of points forming an upward-sloping line.

- Example: Relationship between hours studied and test scores.

```python
# Example data: Hours studied vs. Test scores
np.random.seed(0)
hours_studied = np.random.uniform(1, 10, 50)
test_scores = hours_studied * 5 + np.random.normal(0, 5, 50)

plt.figure(figsize=(10, 5))
plt.scatter(hours_studied, test_scores)
plt.title('Positive Linear Relationship')
plt.xlabel('Hours Studied')
plt.ylabel('Test Scores')
plt.grid(True)
plt.show()
```

####**Negative Linear Relationship**

- In a negative linear relationship, as one variable increases, the other variable decreases. This relationship is depicted as a set of points forming a downward-sloping line.

- Example: Relationship between daily exercise and body fat percentage.

```python
# Example data: Daily exercise (hours) vs. Body fat percentage
np.random.seed(1)
exercise_hours = np.random.uniform(1, 10, 50)
body_fat_percentage = 30 - exercise_hours * 2 + np.random.normal(0, 2, 50)

plt.figure(figsize=(10, 5))
plt.scatter(exercise_hours, body_fat_percentage)
plt.title('Negative Linear Relationship')
plt.xlabel('Daily Exercise (hours)')
plt.ylabel('Body Fat Percentage')
plt.grid(True)
plt.show()
```

####**Non-linear Relationship**

- In a non-linear relationship, the change in one variable does not result in a constant change in the other variable. This can be quadratic, exponential, or any other non-linear form.

- Example: Relationship between speed and braking distance of a car.

```python
# Example data: Speed vs. Braking distance
speed = np.linspace(10, 100, 50)
braking_distance = speed**2 * 0.05 + np.random.normal(0, 10, 50)

plt.figure(figsize=(10, 5))
plt.scatter(speed, braking_distance)
plt.title('Non-linear Relationship')
plt.xlabel('Speed (km/h)')
plt.ylabel('Braking Distance (m)')
plt.grid(True)
plt.show()
```

####**No Relationship**
In a no relationship scenario, the change in one variable does not result in a predictable change in the other variable. The points are scattered randomly on the plot.

Example: Relationship between shoe size and income.

```python
# Example data: Shoe size vs. Income
np.random.seed(3)
shoe_size = np.random.uniform(5, 12, 50)
income = np.random.uniform(20000, 100000, 50)

plt.figure(figsize=(10, 5))
plt.scatter(shoe_size, income)
plt.title('No Relationship')
plt.xlabel('Shoe Size')
plt.ylabel('Income ($)')
plt.grid(True)
plt.show()
```

####**Clustered Data**

- Sometimes data can form clusters indicating different groups or categories within the data. This can be visualized with different colors for each cluster.

- Example: Relationship between age and income, with clusters representing different education levels.

```python
# Example data: Age vs. Income with clusters
np.random.seed(4)
age = np.random.uniform(20, 60, 150)
income = (age * 1000) + np.random.normal(0, 10000, 150)

# Create clusters
education_level = np.random.choice(['High School', 'Bachelor', 'Master', 'PhD'], 150)

plt.figure(figsize=(10, 5))
for level in np.unique(education_level):
    plt.scatter(age[education_level == level], income[education_level == level], label=level)
plt.title('Clustered Data')
plt.xlabel('Age')
plt.ylabel('Income ($)')
plt.legend()
plt.grid(True)
plt.show()
```

###**Customizing Bar Graphs and Pie Charts in Matplotlib**

Matplotlib provides a variety of customization options for bar graphs and pie charts. This note explains how to customize bar graphs by plotting horizontal bars, changing the height and width of the bars, and changing the color of the bars. It also covers customizing pie charts by adding legends, changing the color of each pie slice, exploding slices, adding shadows, and more.



####**Customizing Bar Graphs**

Plotting Horizontal Bar Graph

- Use barh() instead of bar().
Changing the Height and Width of the Bars

- For vertical bars, the width can be adjusted using the width parameter in bar().

- For horizontal bars, the height can be adjusted using the height parameter in barh().
Changing the Color of the Bars

- Use the color parameter to specify colors.

```python
# Example data
categories = ['A', 'B', 'C', 'D']
values = [10, 24, 36, 18]

# Vertical bar graph with customized width and color
plt.figure(figsize=(10, 5))
plt.bar(categories, values, width=0.5, color=['blue', 'green', 'red', 'purple'])
plt.title('Customized Vertical Bar Graph')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.grid(True)
plt.show()

# Horizontal bar graph with customized height and color
plt.figure(figsize=(10, 5))
plt.barh(categories, values, height=0.5, color=['blue', 'green', 'red', 'purple'])
plt.title('Customized Horizontal Bar Graph')
plt.xlabel('Values')
plt.ylabel('Categories')
plt.grid(True)
plt.show()
```

####**Customizing Pie Charts**

- Adding a Legend

Use plt.legend() to add a legend to the pie chart.
Changing the Color of Each Pie Slice

Use the colors parameter to specify colors for each slice.
Exploding Slices

- Use the explode parameter to pull out slices from the pie.
Adding Shadow

- Use the shadow parameter to add a shadow to the pie chart.

```python
import matplotlib.pyplot as plt

# Example data
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.05, 0.09, 0, 0)  # explode 1st slice

plt.figure(figsize=(10, 7))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%',
        shadow=True, startangle=145)
plt.title('Customized Pie Chart')
plt.legend(labels, loc="best")
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
